```python
import pandas as pd 
```

```python
library = pd.read_excel("./library/english_library.xlsx")
```

```python
result = pd.read_excel("../performance/api/aurora-sdg-multi.xlsx")
```

```python
result 
```

```python
def detect(n):
    if n >= 0.5:
        return 1
    else:
        return 0
```

```python
for i in range(1, 18):
    library[i] = library[i].apply(lambda x: detect(x))
```

```python
for i in range(1, 17):
    result[i] = result[i].apply(lambda x: detect(x))
```

```python
result 
```

* Accuracy

```python
def acc(n):
    freq = 0
    summary = len(result)
    for j in range(len(result)):
        if result[n][j] == library[n][j]:
            freq += 1
    return freq/summary
```

* True Positive

```python
def tp(n):
    freq = 0
    for j in range(len(result)):
        if result[n][j] == 1 and library[n][j] == 1:
            freq += 1 
    return freq
```

* False Negative

```python
def fn(n):
    freq = 0
    for j in range(len(result)):
        if result[n][j] == 0 and library[n][j] == 1:
            freq += 1 
    return freq
```

* False Positive

```python
def fp(n):
    freq = 0
    for j in range(len(result)):
        if result[n][j] == 1 and library[n][j] == 0:
            freq += 1 
    return freq
```

* True Negative

```python
def tn(n):
    freq = 0
    for j in range(len(result)):
        if result[n][j] == 0 and library[n][j] == 0:
            freq += 1 
    return freq
```

```python
def precision(tp, fp):
    return tp/(tp+fp)
```

```python
def recall(tp, fn):
    return tp/(tp+fn)
```

```python
def f1(tp, fn, fp):
    return 2*tp/(2*tp+fn+fp)
```

```python
performance = pd.read_excel('./result/performance-table.xlsx')
```

```python
performance 
```

```python
for i in range(0, 16):
    performance["True Positive"][i] = tp(i+1)
    performance["False Negative"][i] = fn(i+1)
    performance["False Positive"][i] = fp(i+1)
    performance["True Negative"][i] = tn(i+1)
    performance["precision"][i] = precision(tp=tp(i+1), fp=fp(i+1))
    performance["recall"][i] = recall(tp=tp(i+1), fn=fn(i+1))
    performance["F1"][i] = f1(tp=tp(i+1),fn=fn(i+1),fp=fp(i+1))
    performance["Acc"][i] = acc(i+1)
```

```python
performance
```
