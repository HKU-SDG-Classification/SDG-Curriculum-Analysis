# draw curves

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score

methods = {
    'Aurora': aurora,
    'AuroraMulti': auroramulti,
    'Elsevier': elsevier,
    'AutoGluon': autogluon
}

fig, axs = plt.subplots(4, 4, figsize=(20, 20))

for sdg in range(1, 17):
    i = (sdg - 1) // 4
    j = (sdg - 1) % 4

    for method, df in methods.items():
        fpr, tpr, _ = roc_curve(ground_truth[f'sdg_{sdg}'], df[f'sdg_{sdg}'])
        roc_auc = auc(fpr, tpr)
        axs[i, j].plot(fpr, tpr, label=f'{method} (AUC = {roc_auc:.2f})')

    axs[i, j].set_title(f'SDG {sdg}')
    axs[i, j].set_xlabel('False Positive Rate')
    axs[i, j].set_ylabel('True Positive Rate')
    axs[i, j].legend()
plt.subplots_adjust(top=0.95)
fig.suptitle('ROC Curves for 16 SDGs')

plt.savefig("ROCCurves.jpg",dpi=300)
plt.show()


fig, axs = plt.subplots(4, 4, figsize=(20, 20))

for sdg in range(1, 17):
    i = (sdg - 1) // 4
    j = (sdg - 1) % 4

    for method, df in methods.items():
        precision, recall, _ = precision_recall_curve(ground_truth[f'sdg_{sdg}'], df[f'sdg_{sdg}'])
        avg_precision = average_precision_score(ground_truth[f'sdg_{sdg}'], df[f'sdg_{sdg}'])
        axs[i, j].plot(recall, precision, label=f'{method} (AP = {avg_precision:.2f})')

    axs[i, j].set_title(f'SDG {sdg}')
    axs[i, j].set_xlabel('Recall')
    axs[i, j].set_ylabel('Precision')
    axs[i, j].legend()
plt.subplots_adjust(top=0.95)
fig.suptitle('P-R Curves for 16 SDGs')

plt.savefig("P-RCurves.jpg",dpi=300)
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, average_precision_score

methods = {
    'Aurora': aurora,
    'AuroraMulti': auroramulti,
    'Elsevier': elsevier,
    'AutoGluon': autogluon
}

auc_scores = {}
ap_scores = {}

for method, df in methods.items():
    auc_scores[method] = []
    ap_scores[method] = []
    for sdg in range(1, 17):
        auc_scores[method].append(roc_auc_score(ground_truth[f'sdg_{sdg}'], df[f'sdg_{sdg}']))
        ap_scores[method].append(average_precision_score(ground_truth[f'sdg_{sdg}'], df[f'sdg_{sdg}']))

    mean_auc = sum(auc_scores[method]) / len(auc_scores[method])
    mean_ap = sum(ap_scores[method]) / len(ap_scores[method])

    print(f'{method}: Mean AUC = {mean_auc:.2f}, Mean AP = {mean_ap:.2f}')
```



# bar chart



```python
# Set the threshold values
thresholds = {"aurora": 0.98, "auroramulti": 0.35, "elsevier": 0.35, "autogluon": 0.5}

# Convert the values to binary outputs based on the thresholds
for model_name, model_df in [("aurora", aurora), ("auroramulti", auroramulti), ("elsevier", elsevier), ("autogluon", autogluon)]:
    threshold = thresholds[model_name]
    model_df[model_df >= threshold] = 1
    model_df[model_df < threshold] = 0
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# Initialize subdataframes for each method
aurora_eval = pd.DataFrame(columns=["precision", "recall", "F1", "accuracy"])
auroramulti_eval = pd.DataFrame(columns=["precision", "recall", "F1", "accuracy"])
elsevier_eval = pd.DataFrame(columns=["precision", "recall", "F1", "accuracy"])
autogluon_eval = pd.DataFrame(columns=["precision", "recall", "F1", "accuracy"])
osdg_eval = pd.DataFrame(columns=["precision", "recall", "F1", "accuracy"])
# Calculate evaluation metrics for each SDG column
for sdg_col in ground_truth.columns:
    y_true = ground_truth[sdg_col]
    
    y_pred_aurora = aurora[sdg_col]
    y_pred_auroramulti = auroramulti[sdg_col]
    y_pred_elsevier = elsevier[sdg_col]
    y_pred_autogluon = autogluon[sdg_col]
    y_pred_osdg = osdg[sdg_col]
    # Calculate evaluation metrics for each method
    aurora_eval.loc[sdg_col] = [
        precision_score(y_true, y_pred_aurora),
        recall_score(y_true, y_pred_aurora),
        f1_score(y_true, y_pred_aurora),
        accuracy_score(y_true, y_pred_aurora),
    ]
    
    auroramulti_eval.loc[sdg_col] = [
        precision_score(y_true, y_pred_auroramulti),
        recall_score(y_true, y_pred_auroramulti),
        f1_score(y_true, y_pred_auroramulti),
        accuracy_score(y_true, y_pred_auroramulti),
    ]
    
    elsevier_eval.loc[sdg_col] = [
        precision_score(y_true, y_pred_elsevier),
        recall_score(y_true, y_pred_elsevier),
        f1_score(y_true, y_pred_elsevier),
        accuracy_score(y_true, y_pred_elsevier),
    ]
    
    autogluon_eval.loc[sdg_col] = [
        precision_score(y_true, y_pred_autogluon),
        recall_score(y_true, y_pred_autogluon),
        f1_score(y_true, y_pred_autogluon),
        accuracy_score(y_true, y_pred_autogluon),
    ]

    osdg_eval.loc[sdg_col] = [
        precision_score(y_true, y_pred_osdg),
        recall_score(y_true, y_pred_osdg),
        f1_score(y_true, y_pred_osdg),
        accuracy_score(y_true, y_pred_osdg),
    ]
# Calculate mean evaluation metrics for each method across all SDGs
aurora_mean = aurora_eval.mean()
auroramulti_mean = auroramulti_eval.mean()
elsevier_mean = elsevier_eval.mean()
autogluon_mean = autogluon_eval.mean()
osdg_mean = osdg_eval.mean()
# Print the mean evaluation metrics
print("Aurora Mean Evaluation Metrics:\n", aurora_mean)
print("\nAurora Multi Mean Evaluation Metrics:\n", auroramulti_mean)
print("\nElsevier Mean Evaluation Metrics:\n", elsevier_mean)
print("\nAutoGluon Mean Evaluation Metrics:\n", autogluon_mean)
print("\nOSDG Mean Evaluation Metrics:\n", osdg_mean)
```
